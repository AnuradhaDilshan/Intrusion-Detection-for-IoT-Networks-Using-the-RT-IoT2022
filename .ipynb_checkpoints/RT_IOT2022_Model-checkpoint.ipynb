```python
# Function to simplify correlation matrix
def corr_simple(data, sig=0.5):
    # Convert object columns (non-numeric) to numeric by converting them to category codes
    dataset_cor = data.copy()
    dataset_cor = dataset_cor.apply(lambda x: x.astype('category').cat.codes if x.dtype == 'object' else x)

    # Compute the correlation matrix
    corr = dataset_cor.corr()

    # Set the lower triangle and diagonal to NaN (to drop perfect correlations and duplicates)
    corr.values[np.tril_indices_from(corr, k=0)] = np.nan

    # Flatten the matrix into a 3-column table (variable 1, variable 2, correlation)
    corr_table = corr.stack().reset_index()
    corr_table.columns = ['Var1', 'Var2', 'Correlation']

    # Drop correlations below the significance level
    corr_filtered = corr_table[corr_table['Correlation'].abs() > sig].sort_values(by='Correlation', ascending=False)

    # Display the filtered correlations
    print(corr_filtered)

    # Convert back to a correlation matrix for visualization
    mtx_corr = corr_filtered.pivot(index='Var1', columns='Var2', values='Correlation')

    # Plot the correlation matrix
    plt.figure(figsize=(15, 10))
    sns.heatmap(mtx_corr, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, linewidths=0.5)
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.show()

# Applying it to dataset
corr_simple(dataset, sig=0.95)
```

```python
# Function to simplify correlation matrix
def corr_simple(data, sig=0.5):
    # Convert object columns (non-numeric) to numeric by converting them to category codes
    dataset_cor = data.copy()
    dataset_cor = dataset_cor.apply(lambda x: x.astype('category').cat.codes if x.dtype == 'object' else x)

    # Compute the correlation matrix
    corr = dataset_cor.corr()

    # Set the lower triangle and diagonal to NaN (to drop perfect correlations and duplicates)
    corr.values[np.tril_indices_from(corr, k=0)] = np.nan

    # Flatten the matrix into a 3-column table (variable 1, variable 2, correlation)
    corr_table = corr.stack().reset_index()
    corr_table.columns = ['Var1', 'Var2', 'Correlation']

    # Drop correlations below the significance level
    corr_filtered = corr_table[corr_table['Correlation'].abs() > sig].sort_values(by='Correlation', ascending=False)

    # Display the filtered correlations
    print(corr_filtered)

    # Get unique variables involved in significant correlations
    significant_vars = pd.unique(corr_filtered[['Var1', 'Var2']].values.ravel('K'))

    return significant_vars

# Applying it to your dataset and getting significant columns
significant_columns = corr_simple(dataset, sig=0.95)

# Plot feature distributions for significant columns only
dataset[significant_columns].hist(figsize=(20, 15), bins=30)
plt.tight_layout()
plt.show()
```

```python
# Drop rows with missing values
dataset.dropna(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# Check if 'Attack_type' exists and label encode it
if 'Attack_type' in dataset.columns:
    le = LabelEncoder()
    dataset['Attack_type'] = le.fit_transform(dataset['Attack_type'])
else:
    print("'Attack_type' column is not in the DataFrame")

# Check if 'proto' and 'service' exist before one-hot encoding
# If they exist, perform one-hot encoding
columns_to_encode = ['proto', 'service']
for col in columns_to_encode:
    if col in dataset.columns:
        dataset = pd.get_dummies(dataset, columns=[col], drop_first=True)
    else:
        print(f"'{col}' column is not in the DataFrame")
```

```python
dataset.head()
```

```python
# Normalize or standardize numerical columns to ensure they are on the same scale.
from sklearn.preprocessing import StandardScaler
columns_to_exclude = []  # Initial list

# Dynamically find any columns that start with 'proto' or 'service'
proto_columns = dataset.filter(like='proto').columns
service_columns = dataset.filter(like='service').columns

# Step 1: Add these to the exclude list
columns_to_exclude.extend(proto_columns)
columns_to_exclude.extend(service_columns)

# Step 2: Drop these columns or keep them separate before scaling
dataset_scaled = dataset.drop(columns=columns_to_exclude, errors='ignore')

# Step 3: Identify all remaining numerical columns
numerical_cols = dataset_scaled.select_dtypes(include=['float64', 'int64']).columns

# Step 4: Initialize the StandardScaler
scaler = StandardScaler()

# Step 5: Fit and transform the numerical columns
dataset_scaled[numerical_cols] = scaler.fit_transform(dataset_scaled[numerical_cols])

# Add back the non-numerical columns (if needed)
dataset_final = pd.concat([dataset_scaled, dataset[columns_to_exclude]], axis=1)

# Display the first few rows of the correctly scaled data
print(dataset_final.head())
```

### Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (classification_report, confusion_matrix, f1_score, accuracy_score, precision_score, recall_score)
from tqdm import tqdm  # Import tqdm for progress bar

# Encode the target variable since it's categorical
le = LabelEncoder()
y = le.fit_transform(dataset_final['Attack_type'])

# Define features and target variable
X = dataset_final.drop(['Attack_type'], axis=1)

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize models
models = {
    'Random Forest Classifier': RandomForestClassifier(n_estimators=100, random_state=42),
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Support Vector Machine': SVC(random_state=42)
}

# Lists to store results
model_names = []
accuracies = []
precisions = []
recalls = []
f1_scores = []

# Train and evaluate each model
for name, model in models.items():
    print(f"\nTraining {name}...")
    model.fit(X_train, y_train)
    print(f"Finished training {name}.")
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    
    # Store results
    model_names.append(name)
    accuracies.append(accuracy)
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1)

# Create results DataFrame
results = pd.DataFrame({
    'Model': model_names,
    'Accuracy': accuracies,
    'Precision': precisions,
    'Recall': recalls,
    'F1 Score': f1_scores
})

# Choose the metric to determine the best model
metric_to_optimize = 'F1 Score'  # You can change this to 'Accuracy', 'Precision', or 'Recall'

# Find the best model based on the selected metric
best_model_row = results.loc[results[metric_to_optimize].idxmax()]
best_model_name = best_model_row['Model']
best_model = models[best_model_name]

# Display the results
print("\nModel Performance Metrics:")
print(results)
print(f"\nBest Model based on {metric_to_optimize}: {best_model_name}")
```

### Inference Pipeline

```python
def make_prediction(model, input_data):
    """
    Make predictions using a trained model
    
    Args:
        model: Trained ML model
        input_data: Features to predict on (should be preprocessed the same way as training data)
        
    Returns:
        Prediction from the model
    """
    # Get the feature names from the model
    if hasattr(model, 'feature_names_in_'):
        required_features = model.feature_names_in_
    else:
        # If model doesn't expose feature names, we'll need to ensure all features are present
        print("Warning: Model doesn't expose feature names. Ensure input has all required features.")
        
    # Ensure input data has same features as training data
    if isinstance(input_data, pd.DataFrame):
        # If DataFrame, ensure it has all required features
        features = input_data.copy()
    else:
        # If single sample, convert to DataFrame with all required features
        features = pd.DataFrame([input_data])
    
    # If we have the required feature names, ensure they match
    if hasattr(model, 'feature_names_in_'):
        # Check for missing features
        missing_features = set(required_features) - set(features.columns)
        if missing_features:
            print(f"Adding missing features: {missing_features}")
            for feature in missing_features:
                features[feature] = 0  # Default value for missing features
        
        # Ensure columns are in the same order as during training
        features = features[required_features]
    
    # Make prediction
    prediction = model.predict(features)
    
    return prediction

# Example usage:
# Choose your best performing model
best_model = models['Random Forest Classifier'] # Or whichever model performed best

# Example input data with all required features
sample_input = {
    'flow_duration': 32.0,
    'fwd_pkts_tot': 9,
    'bwd_pkts_tot': 5,
    'fwd_data_pkts_tot': 3,
    'bwd_data_pkts_tot': 2,
    'fwd_pkts_per_sec': 0.28,
    'bwd_pkts_per_sec': 0.16,
    'flow_pkts_per_sec': 0.44,
    'active.min': 0,
    'active.max': 0,
    'active.avg': 0,
    'active.std': 0,
    'active.tot': 0,
    'idle.min': 0,
    'idle.max': 0,
    'idle.avg': 0,
    'idle.std': 0,
    'fwd_init_window_size': 64240,
    'bwd_init_window_size': 26847,
    'fwd_last_window_size': 502,
    'bwd_PSH_flag_count': 0,
    'bwd_URG_flag_count': 0,
    'bwd_bulk_bytes': 0,
    'bwd_bulk_packets': 0,
    'bwd_bulk_rate': 0,
    'service_dns': False,
    'service_http': False,
    'service_irc': False,
    'service_mqtt': True,
    'service_ntp': False,
    'service_radius': False,
    'service_ssh': False,
    'service_ssl': False,
    # Adding missing features mentioned in the error
    'bwd_header_size_max': 0,
    'bwd_header_size_min': 0,
    'bwd_header_size_tot': 0,
    'bwd_iat.avg': 0,
    'bwd_iat.max': 0
    # Note: There may be more missing features not shown in the error message
}

# Make prediction
try:
    prediction = make_prediction(best_model, sample_input)
    print(f"\nPredicted Attack Type: {prediction[0]}")
except Exception as e:
    print(f"Error making prediction: {e}")
    print("Please ensure all features used during training are present in the input data.")
```

```python
# Use the correct LabelEncoder for 'Attack_type' from the dictionary
try:
    attack_type_encoder = label_encoders['Attack_type']
    predicted_label = attack_type_encoder.inverse_transform([int(prediction[0])])[0]
    print(f"Predicted Attack Type Label: {predicted_label}")
except Exception as e:
    print(f"Error decoding prediction label: {e}")
```
